```python
x=df.drop(['price','ID'],axis=1)
x
```

```python
y=df['price']
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x['name']=le.fit_transform(x['name'])
x['fuelsystem']=le.fit_transform(x['fuelsystem'])
x['aspiration']=le.fit_transform(x['aspiration'])
x['doornumbers']=le.fit_transform(x['doornumbers'])
x['carbody']=le.fit_transform(x['carbody'])
x['drivewheels']=le.fit_transform(x['drivewheels'])
x['cylindernumber']=le.fit_transform(x['cylindernumber'])
x['fueltypes']=le.fit_transform(x['fueltypes'])
x['enginetype']=le.fit_transform(x['enginetype'])
x['enginelocation']=le.fit_transform(x['enginelocation'])
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
s.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
R2 = r2_score(y_test,y_pred)
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
print('R squared is',R2,'\n','MAE is',MAE,'\n','MSE is',MSE,'\n','RMSE is',RMSE)
```
